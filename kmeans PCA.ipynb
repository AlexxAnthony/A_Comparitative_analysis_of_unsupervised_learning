```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("riceClassification.csv")
df
```

```python
df.drop(["id","Class"],axis=1,inplace=True)
df
```

```python
plt.figure(figsize=(20,20))
plt.title("Heatmap Corralation")
sns.heatmap(data=pd.get_dummies(df).corr(),annot=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
x_pca
```

```python
X=x_pca
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i,random_state = 42)
    kmeans.fit(X)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
y_kmeans
```

```python
df["clusters"]=y_kmeans
df
```

```python
y=y_kmeans
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)

kmeans.fit(X)

# check how many of the samples were correctly labeled
labels = kmeans.labels_

correct_labels = sum(y == labels)
print("Result: %d out of %d samples were correctly labeled." % (correct_labels, y.size))
print('Accuracy score: {0:0.2f}'. format(correct_labels/float(y.size)))
```

```python
import numpy as np
import matplotlib.pyplot as plt


plt.scatter(X[labels == 0, 0], X[labels == 0, 1], label='Class 0', c='blue')
plt.scatter(X[labels == 1, 0], X[labels == 1, 1], label='Class 1', c='red')

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of PCA Components with Different Colors')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
kmeans.cluster_centers_
```

```python
cluster_assignments = kmeans.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=cluster_assignments, cmap='viridis', edgecolors='k', s=50)
# 'c' parameter assigns colors based on cluster_assignments
# 'cmap' parameter specifies the color map
# 'edgecolors' parameter sets the color of the marker edges
# 's' parameter controls the marker size

# Plot the cluster centers
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='X', s=200, alpha=0.75)
# 'c' parameter sets the color to red
# 'marker' parameter specifies the marker shape (X)
# 's' parameter controls the marker size
# 'alpha' parameter controls the transparency

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-means Clustering with Colored Cluster Assignments')

# Show the plot
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
cluster_assignments = kmeans.fit_predict(X)
```

```python
silhouette_avg = silhouette_score(X, cluster_assignments)
print(f"Silhouette Score (K-means): {silhouette_avg}")
```
